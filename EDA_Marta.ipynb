```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lee el archivo CSV
data = pd.read_csv('HR_RAW_DATA.csv')
```

```python
def exploracion_basic_datos(df):
    # Información general del DataFrame
    print('Información general del DataFrame:\n')
    print(df.info())
    print('________________________________________\n')

    # Mostrar todas las columnas
    pd.set_option('display.max_columns', None)
    print('Las cinco primeras filas del DataFrame sin puntos suspensivos:\n')
    print(df.head())
    print('________________________________________\n')

    # Forma del DataFrame
    print('La forma del DataFrame es:')
    print(f"El número de filas es {df.shape[0]}, y el número de columnas es {df.shape[1]}\n")
    print('________________________________________\n')

    # Estadísticas descriptivas transpuestas
    print('El descriptivo del DataFrame incluyendo todas las columnas es:\n')
    print(df.describe(include='all').T)
    print('________________________________________\n')

    # Tipos de datos por columna
    print('El tipo de dato que contiene cada columna es de:\n')
    print(df.dtypes)
    print('________________________________________\n')

    # Valores nulos por columna
    print('La cantidad de nulos por columna es de:\n')
    print(df.isnull().sum())
    print('________________________________________\n')

    # Valores duplicados
    print(f"El DataFrame tiene {df.duplicated().sum()} filas duplicadas.")
    print('________________________________________\n')
```

```python
exploracion_basic_datos(data)
```

```python
def exploracion_datos_adv(df):
    # Convertir valores no numéricos a NaN para variables numéricas
    for column in df.select_dtypes(include=['object']).columns:
        try:
            df[column] = pd.to_numeric(df[column], errors='coerce')
        except ValueError:
            pass

    # Histograma de las variables numéricas
    print("Histograma de las variables numéricas:")
    try:
        df.hist(figsize=(15, 10), bins=30)
        plt.tight_layout()
        plt.show()
    except ValueError as e:
        print(f"Error al generar histogramas: {e}")

    # Gráfico de correlación
    print("Matriz de correlación:")
    try:
        plt.figure(figsize=(12, 8))
        sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
        plt.show()
    except ValueError as e:
        print(f"Error al generar la matriz de correlación: {e}")

    # Boxplots de las variables numéricas
    print("Boxplots de las variables numéricas:")
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        try:
            plt.figure(figsize=(10, 6))
            sns.boxplot(x=df[column])
            plt.title(f'Boxplot de {column}')
            plt.show()
        except ValueError as e:
            print(f"Error al generar boxplot para {column}: {e}")
```

```python
exploracion_datos_adv(data)
```

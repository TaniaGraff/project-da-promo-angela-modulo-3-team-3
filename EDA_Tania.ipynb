```python
#Creamos la funciÃ³n para explorar los datos.
def exploracion_datos(df):
    print('InformaciÃ³n general del DataFrame:\n')
    print(df.info())
    print('________________________________________\n')

    print(pd.set_option('display.max_columns', None))
    print('Las cinco primeras filas del DataFrame sin puntos suspensivos:\n')
    print(df.head())
    print('________________________________________\n')

    print('La forma del DataFrame es:')
    print(f"El nÃºmero de filas que tenemos es de {df.shape[0]}, y el nÃºmero de columnas es de {df.shape[1]}\n")
    print('________________________________________\n')

    print('El descriptivo del DataFrame incluyendo todas las columnas es:\n')
    print(df.describe(include='all'))
    print('________________________________________\n')

    print('El tipo de dato que contiene cada columna es de:\n')
    print(df.dtypes)
    print('________________________________________\n')

    print('La cantidad de nulos por columna es de:\n')
    print(df.isnull().sum())
    print('________________________________________\n')

    print('La cantidad de valores duplicados por columna es de:\n')
    print(df.duplicated().sum())
 

```

```python
exploracion_datos(df_HR)
```

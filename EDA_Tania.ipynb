```python
#Creamos la funciÃ³n para explorar los datos.
def exploracion_datos(df):
    print('___________ INFORMACIÃ“N GENERAL DEL DATAFRAME _________\n')
    print(df.info())

    print('___________________ FORMA DEL DATAFRAME ________________\n')
    
    print(f"El nÃºmero de filas que tenemos es de {df.shape[0]}.\nEl nÃºmero de columnas es de {df.shape[1]}\n")
    

    print('___________________ NULOS Y DUPLICADOS __________________\n')
    
    print('La cantidad de valores NULOS por columna es de:\n')
    print(df.isnull().sum())
    print('_________________________________________________________\n')

    print(f'La cantidad de valores duplicados por columna es de: {df.duplicated().sum()}\n')

    print('___________________ RESUMEN ESTADÃSTICO ________________')
    print('___________________ Variables NumÃ©ricas ________________\n')
    print(df.describe().T)
    
    print('__________________ Variables CategÃ³ricas _______________\n')
    print(df.describe(include='object').T)
 
```

```python
exploracion_datos(df_HR)
```

```python
#FASE 2
#Reemplazamos los valores 1 y 0 de la columna Gender por "Male" y "Female", o "M" y "F" por ejemplo, usando loc.

print(df_HR['Gender'])

df_HR.loc[df_HR['Gender'] == 0, 'Gender'] = 'M'
df_HR.loc[df_HR['Gender'] == 1, 'Gender'] = 'F'
```

```python
#Â¡Â¡Â¡OJO!!! HACER UNA COPIA DEL DATA FRAME POR SI ACASO. NO EJECUTO CELDA PUES ME DA ERROR PORQUE LOS DATOS YA SON FLOAT.
print(df_HR['DailyRate'].dtype)
print(df_HR['DailyRate'])

#Reemplazamos comas por puntos en los valores de la columna 'DailyRate'
df_HR['DailyRate'] = df_HR['DailyRate'].str.replace(',', '.')

#Eliminamos todos los caracteres que no sean dÃ­gitos o puntos de los valores de la columna 'DailyRate'
df_HR['DailyRate'] = df_HR['DailyRate'].str.replace('[^\d\.]', '', regex=True)

#Reemplazamos los valores vacÃ­os por 'NaN'. Â¡Â¡Â¡OJO!!! IGUAL NOS CONVIENE MÃS REMPLAZARLO POR 0.0. REVISAR
df_HR['DailyRate'] = df_HR['DailyRate'].replace('', 'NaN')

#Convertimos la columna 'DailyRate' a tipo float
df_HR['DailyRate'] = df_HR['DailyRate'].astype(float)
```
